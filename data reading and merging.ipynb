```python
import mysql.connector
import getpass
import pandas as pd
```

```python
cnx = mysql.connector.connect(host='localhost', port = 3306, database = "ih_gambling", user = "root", password = password)
```

```python
cursor = cnx.cursor()
```

```python
cursor.execute("SHOW TABLES") #to see what tables are in schema
print(cursor.fetchall())
```

```python
query = ("SELECT * FROM account")
cursor.execute(query)
results = cursor.fetchall()
columns = ["AccountNo", "CustId", "AccountLocation", "CurrencyCode", "DailyDepositLimit", "StakeScale", "SourceProd"]
account_df = pd.DataFrame(results, columns=columns)
account_df = account_df.drop(columns=["AccountLocation"])
account_df #10 accounts
```

```python
query = ("SELECT * FROM customer")
cursor.execute(query)
customer = cursor.fetchall()
columns = ["CustId", "AccountLocation", "Title", "FirstName", "LastName", "CreateDate", "CountryCode", "Language", "Status", "DateOfBirth", "Contact", "CustomerGroup"]
customer_df = pd.DataFrame(customer, columns=columns)
customer_df #10 accounts
```

```python
customer_df = pd.merge(left=account_df, right=customer_df, left_on="CustId", right_on="CustId")
customer_df
```

```python
query = ("SELECT * FROM product")
cursor.execute(query)
columns = [desc[0] for desc in cursor.description] #fetches column names from the cursor
product = cursor.fetchall()

product_df = pd.DataFrame(product, columns=columns)
product_df #46 products
```

```python
query = ("SELECT * FROM betting")
cursor.execute(query)
columns = [desc[0] for desc in cursor.description] #fetches column names from the cursor

betting = cursor.fetchall() 

betting_df = pd.DataFrame(betting, columns=columns)
betting_df = betting_df.drop(columns=["MyUnknownColumn", "MyUnknownColumn_[0]", "MyUnknownColumn_[1]", "MyUnknownColumn_[2]", "AccountNo_[1]", "Product_[0]"])
# what is with Bet_Amt_[0]?
betting_df #630 rows

```

```python
#merging
merged_df_1 = pd.merge(left=betting_df, right=customer_df, left_on="AccountNo", right_on="AccountNo", how="left")
merged_df = pd.merge(left=merged_df_1, right=product_df, left_on="ClassId", right_on="CLASSID", how="left")
merged_df.to_csv("merged_df.csv")
```

```python
#analysis ideas
# Customer Behavior & Segmentation
#Top Customers by Bet Amount (Bet_Amt): Who are the “high rollers”?
#Behavior by CustomerGroup (Gold, Silver, Bronze): Do Gold customers actually bet more?
#Analysis by Age Groups (derivable from DateOfBirth): Who bets more – younger or older users
# 
#Profitability Analysis
#house Profit = Bet_Amt - Win_Amt
#Who loses the most?
#Which products are the most profitable?
#Filter customers with negative vs. positive winnings – Who are the heavy players that often lose?

df = pd.read_csv("merged_df.csv")

# dropping further columns:
# col unnamed: 0 is an index that we dont need further in the analysis. 
# col product, classid and categoryid were doubled 
# col Status is always an A
# col accountloaction is always GIB
df = df.drop(columns=["Unnamed: 0", "product", "Status", "CLASSID", "CATEGORYID", 'Bet_Amt_[0]',
       'Vegas', 'Sportsbook', 'Games', 'Casino', 'Poker', 'Bingo', '#N/A']) 

#try to find out why 2 columns for account number
df["AccountNo match?"] = df["AccountNo"] == df["AccountNo_[0]"]
df.columns

#convert the currency values all to EUR
def convert_to_eur_bet(row):
    if row["CurrencyCode"] == "EUR":
        return row["Bet_Amt"] * 1.00
    elif row["CurrencyCode"] == "USD":
        return row["Bet_Amt"] * 0.93
    elif row["CurrencyCode"] == "GBP":
        return row["Bet_Amt"] * 1.17
    else:
        return None
    
df["Bet_Amt_EUR"] = df.apply(convert_to_eur_bet, axis =1)
df

def convert_to_eur_win(row):
    if row["CurrencyCode"] == "EUR":
        return row["Win_Amt"] * 1.00
    elif row["CurrencyCode"] == "USD":
        return row["Win_Amt"] * 0.93
    elif row["CurrencyCode"] == "GBP":
        return row["Win_Amt"] * 1.17
    else:
        return None
    
df["Win_Amt_EUR"] = df.apply(convert_to_eur_win, axis =1)

#create a col for full name
df["Full Name"] = df["FirstName"] + " " + df["LastName"]
df["Won the bet?"] = df["Win_Amt_EUR"] != 0
df

#df[df["bet_or_play"] == 0] only 1 row with value of 0 >> adjustments?
```
